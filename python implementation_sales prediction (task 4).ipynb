# step 6 : Model Creation

```python
data.head()
```

```python
# Independent and dependent variable creation
X=data[['TV','Radio','Newspaper']] ## X variable contains all inputs
y=data.Sales ## y is output
```

```python
y
```

```python
# Traning and testing data creation
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
y_train
```

# 1.linear regression

```python
# Creating model
from sklearn.linear_model import LinearRegression
LR=LinearRegression() 
LR.fit(X_train,y_train) 
y_predict=LR.predict(X_test)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
y_test # Actual values
```

```python
y_predict # Values predicted by model
```

```python
#  Model Evalution
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2score=r2_score(y_test,y_predict)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math


print("R2 Score is:", r2score)

# Calculate adjusted R2 score 
adjusted_r2 = 1 - (1 - r2score) * (40 - 1) / (40 - 3 - 1)
print("Adjusted R2:", adjusted_r2)

# Calculate mean squared error and root mean squared error
mse = mean_squared_error(y_test, y_predict)
print("Mean Squared Error:", mse)

rmse = math.sqrt(mse)
print("Root Mean Squared Error:", rmse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_predict)
print("Mean Absolute Error:", mae)
```

# 2.decision tree 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
# Initialize the Decision Tree Regressor
dt_model = DecisionTreeRegressor(max_depth=4, random_state=42)  

# Fit the model 
dt_model.fit(X_train, y_train)

```

```python
# Predict on the test set
y_pred = dt_model.predict(X_test)
```

```python
r2score=r2_score(y_test,y_pred)
print("R2 Score is:", r2score)
# Calculate adjusted R2 score 
adjusted_r2 = 1 - (1 - r2score) * (40 - 1) / (40 - 3 - 1)
print("Adjusted R2:", adjusted_r2)

# Calculate mean squared error and root mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

rmse = math.sqrt(mse)
print("Root Mean Squared Error:", rmse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

# 3. Random forest

```python
from sklearn.ensemble import RandomForestRegressor
# Initialize the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=5)  

# Fit the model 
rf_model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_prediction = rf_model.predict(X_test)
```

```python
r2score=r2_score(y_test,y_prediction)
print("R2 Score is:", r2score)
# Calculate adjusted R2 score 
adjusted_r2 = 1 - (1 - r2score) * (40 - 1) / (40 - 3 - 1)
print("Adjusted R2:", adjusted_r2)

# Calculate mean squared error and root mean squared error
mse = mean_squared_error(y_test, y_prediction)
print("Mean Squared Error:", mse)

rmse = math.sqrt(mse)
print("Root Mean Squared Error:", rmse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_prediction)
print("Mean Absolute Error:", mae)
```
