# step 4 : feature Engineering

```python
# 1.cheking for null values
data.isnull().sum() # there are null  value present in the column age,cabin and embarked
```

```python
# null value handling
data['Age'] = data['Age'].fillna(data['Age'].median()) # filled with median 
data = data.drop(columns=['Cabin']) # majority of value is missing so drop this column
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0]) # since most of the passenger embarked from S

```

```python
data.head(10)
```

```python
data.isnull().sum() # after handling missing values there is no null values left
```

```python
# 2. Duplicate data checking
data[data.duplicated()] # all rows contain duplicate data
```

```python
# handle duplicates
data.drop_duplicates(inplace=True) # simply drop duplicate rows 
```

```python
# as we analyse the data we observed that columns like name, age,ticket and cabin does not affect on target variable there  it is good to drop this column for better prediction
data = data.drop(columns=['PassengerId', 'Ticket', 'Name'])

print(data.head())
```

```python
# 3. checking for outlier present
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Create the boxplot for all numerical columns
plt.figure(figsize=(15, 10))  
sns.boxplot(data=numerical_data)
plt.xticks(rotation=90)  
plt.title('Boxplot of All Numerical Columns')
plt.show()
```

```python
data.columns
```

```python

object_data = data.select_dtypes(include=['object'])

# Plot countplot for each object column
for column in object_data.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=data, x=column)
    plt.xticks(rotation=45)  
```

```python
from scipy import stats

# Calculate the Z-scores for all numerical columns
z_scores = stats.zscore(data.select_dtypes(include=['float64', 'int64']))

# Set a threshold for Z-scores (typically 3 or -3)
threshold = 3

# Remove rows with Z-scores above the threshold (outliers)
data_no_outliers = data[(abs(z_scores) < threshold).all(axis=1)]

# Check the new shape of the DataFrame
print(data_no_outliers.shape)
```

```python

# Boxplot after removing outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=data_no_outliers)
plt.title('Boxplot After Removing Outliers')
plt.xticks(rotation=90)
plt.show()
```

```python
## still there is outlier prrsent in the column age nd fare
# Calculate Q1 and Q3 for 'Age' and 'Fare'
# Calculate Q1 and Q3 for 'Age' and 'Fare'
Q1_age = data['Age'].quantile(0.25)
Q3_age = data['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

Q1_fare = data['Fare'].quantile(0.25)
Q3_fare = data['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare

# Define outlier thresholds for Age and Fare
age_lower = Q1_age - 1.5 * IQR_age
age_upper = Q3_age + 1.5 * IQR_age

fare_lower = Q1_fare - 1.5 * IQR_fare
fare_upper = Q3_fare + 1.5 * IQR_fare

# Remove outliers in Age and Fare based on adjusted thresholds
data_no_outliers = data[
    (data['Age'] >= age_lower) & (data['Age'] <= age_upper) &
    (data['Fare'] >= fare_lower) & (data['Fare'] <= fare_upper)
]

# Check the new shape and visualize again
print(f"Data shape after removing outliers: {data_no_outliers.shape}")
```

```python
## Creating testing and training data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

# 1.Logistic regression

```python

from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train,y_train)  ## training
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

recall = recall_score(y_test, y_pred, average='macro')
print("Recall score is:", recall)

precision = precision_score(y_test, y_pred, average='macro')
print("Precision score is:", precision)

f1 = f1_score(y_test, y_pred, average='macro')
print("F1 score is:", f1)

accuracy = accuracy_score(y_test, y_pred)
print("The accuracy score is:", accuracy)
```

# 2.KNN (k neartest neighbour)

```python
# using knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
error_rate=[]
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
## ploting k value and error rate
plt.figure(figsize=(10,6))
plt.plot(range(1,11),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```
